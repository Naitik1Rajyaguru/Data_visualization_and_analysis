```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
import calendar
import re

# Load the dataset
data = pd.read_csv('traffic_accidents.csv')
```

```python
print(data)
# Calculate correlation matrix
correlation_matrix = data.corr()

# Print correlation matrix
# print(correlation_matrix)

# Find correlations with the target variable ('NUMBER OF PERSONS KILLED')
target_correlations = correlation_matrix['NUMBER OF PERSONS KILLED'].abs().sort_values(ascending=False)
print(target_correlations)
```

```python
# attribute selection

print(data.columns)

# ['CRASH DATE', 'CRASH TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE',
#        'LONGITUDE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME',
#        'OFF STREET NAME', 'NUMBER OF PERSONS INJURED',
#        'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
#        'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
#        'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
#        'NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1',
#        'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
#        'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
#        'COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
#        'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5']

usefull = ['CRASH DATE', 'CRASH TIME', 'LOCATION']
df = data[usefull]
```

```python
# killed/ injured persons

df['PERSON INJURED'] = data[['NUMBER OF CYCLIST INJURED', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF MOTORIST INJURED']].sum(axis=1)
df['PERSON KILLED'] = data[['NUMBER OF CYCLIST KILLED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF MOTORIST KILLED']].sum(axis=1)
df
```

```python
# convert string into class :- emoty values as well.

Labelencoder = LabelEncoder()
df['VEHICLE TYPE'] = Labelencoder.fit_transform(data['VEHICLE TYPE CODE 1'])
df['DRIVER REASON'] = Labelencoder.fit_transform(data['CONTRIBUTING FACTOR VEHICLE 1'])
print(len(pd.unique(data['VEHICLE TYPE CODE 1'])))
print(len(pd.unique(df['VEHICLE TYPE'])))
print(len(pd.unique(data['CONTRIBUTING FACTOR VEHICLE 1'])))
print(len(pd.unique(df['DRIVER REASON'])))
```

```python
# find day

days =["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

def findDay(date):
    month, day, year = (int(i) for i in re.split('/|-', date))
    dayNumber = calendar.weekday(year, month, day)

    return dayNumber

df['DAY'] = data['CRASH DATE'].apply(findDay)
print(df['DAY'])
# df['DAY'] = Labelencoder.fit_transform(df['DAY'])
# print(df['DAY'])
```

```python
# get hour of day

def getTime(str):
    x = [int(i) for i in re.split(':', str)]
    return x[0]
df['HOUR'] = data['CRASH TIME'].apply(getTime)
print(df['HOUR'])
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# Assuming your DataFrame is called 'data' with latitude and longitude columns 'LATITUDE' and 'LONGITUDE'
# data = pd.read_csv('/path/to/your/data.csv')  # Replace with the actual path to your data

# Select the latitude and longitude columns
locations = data[['LATITUDE', 'LONGITUDE']]

# Handle missing values
# locations = locations.dropna()  # Remove rows with missing values

# Fill missing values with a distinct marker
missing_marker = -99999
locations.fillna(missing_marker, inplace=True)

# Define the number of areas (clusters)
num_clusters = 50

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['AREA'] = np.nan  # Initialize the 'AREA' column with NaN
df.loc[locations.index, 'AREA'] = kmeans.fit_predict(locations)  # Assign cluster labels to existing locations

# Assign a separate cluster to missing locations
df.loc[df['AREA'].isnull(), 'AREA'] = num_clusters

# Print the updated DataFrame
print(df['AREA'])
```

```python
# Obtain the cluster centers (k-means)
cluster_centers = kmeans.cluster_centers_

# Print all the k-means
for i in range(num_clusters):
    print(f"Cluster {i} Center: {cluster_centers[i]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Group the data by 'DAY_OF_WEEK' and 'HOUR_OF_DAY' and calculate the sum of 'NUMBER_OF_PERSONS_KILLED'
killed_per_hour = datax.groupby(['DAY', 'HOUR'])['PERSON KILLED'].sum().reset_index()

# Pivot the DataFrame to rearrange the data for plotting
pivoted_data = killed_per_hour.pivot('HOUR', 'DAY', 'PERSON KILLED')

# Set the plot size and style
plt.figure(figsize=(10, 6))
sns.set(font_scale=1.2)

# Create a heatmap plot
ax = sns.heatmap(pivoted_data, cmap='YlOrRd', linewidths=0.5, linecolor='lightgray')

# Set plot title and axis labels
plt.title('Number of Persons Killed per Day and Hour', fontsize=16, fontweight='bold')
plt.xlabel('Day of Week', fontsize=14)
plt.ylabel('Hour of Day', fontsize=14)

# Set x-axis and y-axis tick labels
day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hour_labels = [f'{h:02d}:00' for h in range(24)]
# Set x-axis and y-axis tick labels

ax.set_xticklabels(day_labels)
ax.set_yticklabels(hour_labels)

# Rotate x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add colorbar legend
cbar = ax.collections[0].colorbar
cbar.set_label('Number of Persons Killed', fontsize=12)

# Display the plot
plt.show()

# Save the plot as an image file (e.g., PNG)
plt.savefig('/content/drive/MyDrive/Machine_Learning/KILLED_vs_day_hr_plot.png', dpi=300)

# Display a message indicating the successful save
print("Plot saved as PERSON_KILLED_vs_day_hr_plot.png")
```
